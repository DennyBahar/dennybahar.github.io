```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
sp500 = pd.read_csv("dataset/sp500.csv",index_col='Date',parse_dates=True)
print(sp500.info())
print(sp500.head())
```

```python
sp500['Dividend'] = sp500['Dividend'].astype('float64')
sp500['Dividend'] = sp500['Dividend'].fillna(method='ffill')
print(sp500.info())
```

```python
sp500['annual return'] = 12*sp500['SP500'].pct_change()
sp500['avg return'] = sp500['annual return'].rolling(120).mean()
sp500['dividend rate'] = sp500['Dividend']/sp500['SP500']
sp500['avg dividend'] = sp500['dividend rate'].rolling(120).mean()
sp500['avg interest rate'] = (sp500['Long Interest Rate']/100).rolling(120).mean()

print(sp500.info())
print(sp500.head())
```

```python
ret_arr = sp500['avg return']
div_arr = sp500['avg dividend']
rate_arr = sp500['avg interest rate']
```

```python
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(10,8))

ax[0].plot(ret_arr)

ax[1].plot(div_arr)

ax[2].plot(rate_arr)
```
